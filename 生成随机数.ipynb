```python
import pandas as pd
import numpy as np
```

```python
#生成1到10服从均匀分布的随机数
random_uni=np.random.uniform(1,10,10000)
#生成均值为3.标准差为9的正态分布随机数
random_nor=np.random.normal(3,9,10000)
#生成均值为5的泊松分布随机数
random_poi=np.random.poisson(5,10000)
#生成均值为7的指数分布随机数
random_exp=np.random.exponential(7,10000)
```

```python
#把随机数放进数据框中,通过describe函数输出标准差，均值，最小值
random_uni=pd.DataFrame(random_uni)
random_uni.describe(include='all')
```

```python
random_nor=pd.DataFrame(random_nor)
random_nor.describe()
```

```python
random_poi=pd.DataFrame(random_poi)
random_poi.describe()
```

```python
random_exp=pd.DataFrame(random_exp)
random_exp.describe()
```
